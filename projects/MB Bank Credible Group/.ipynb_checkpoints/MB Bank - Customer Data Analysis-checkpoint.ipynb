### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Prepare the data

```python
df = pd.read_csv('dataset.csv')
df = df.dropna(subset = ['nhomno_xhtdJuin2021'])
df['age'] = 2020 - df.birth_incorp_date
df.drop('birth_incorp_date', inplace = True, axis = 1)
df.rename(columns = lambda x: x.strip(), inplace = True)
df.rename(columns = {'local_ref_1': 'gender', 'vn_marital_status': 'marital_status'}, inplace = True)
df = df.set_index('rd_id')
df
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis', ax = ax)
plt.gcf().text(0.925, 0.5, 'Yellow represents missing values.\n\n Purple represents existing values.', fontsize = 14)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 5))
def make_autopct(values):
    def autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}\n({p:.2f}%)'.format(v = val, p = pct)
    return autopct
    
values = df[df['nhomno_xhtdJuin2021'] < 3].groupby(['gender']).count()['nhomno_xhtdJuin2021'].to_list()
df[df['nhomno_xhtdJuin2021'] < 3].groupby(['gender']).count().plot(kind = 'pie', 
                                                                      y = 'nhomno_xhtdJuin2021', 
                                                                      labels = None, 
                                                                      legend = None, 
                                                                      ylabel = '',
                                                                      autopct = make_autopct(values),
                                                                      ax = ax[0], 
                                                                      startangle = 90)
ax[0].set_xlabel('Credible')

values = df[df['nhomno_xhtdJuin2021'] >= 3].groupby(['gender']).count()['nhomno_xhtdJuin2021'].to_list()
df[df['nhomno_xhtdJuin2021'] >= 3].groupby(['gender']).count().plot(kind = 'pie', 
                                                                    y = 'nhomno_xhtdJuin2021', 
                                                                    labels = None, 
                                                                    ylabel = '', 
                                                                    autopct = make_autopct(values),
                                                                    ax = ax[1], 
                                                                    startangle = 90)
ax[1].set_xlabel('Incredible')

x = np.arange(len(list(df["gender"].unique())[:2]))
width = 0.2
smaller = df[df['nhomno_xhtdJuin2021'] < 3].groupby(['gender']).count()['nhomno_xhtdJuin2021'].to_list()
larger = df[df['nhomno_xhtdJuin2021'] >= 3].groupby(['gender']).count()['nhomno_xhtdJuin2021'].to_list()

y_credible = smaller[::-1]
y_incredible = larger[::-1]

rects1 = ax[2].bar(x - width/2, y_credible, width, label = 'Credible', color = 'g')
rects2 = ax[2].bar(x + width/2, y_incredible, width, label = 'Incredible', color = 'indianred')

ax[2].set_ylabel('Number of Customers', fontsize = 10)
ax[2].set_title('Number of Customers in credible and incredible group categorized by gender', fontsize = 10)

plt.xticks(x, list(df["gender"].unique())[:2], fontsize = 10)
plt.yticks(fontsize = 10)
ax[2].legend(fontsize = 10)

ax[2].bar_label(rects1, padding = 3, fontsize = 10)
ax[2].bar_label(rects2, padding = 3, fontsize = 10)

plt.tight_layout()
```

## <center> Model <center>

## 1. Prepare the data

### 1.1 Remove age outlier

```python
sns.boxplot(y = 'age', data = df)
```

```python
df = df[df['age'] < 100]
```

### 1.2. Fill missing values

```python
df.age = df.age.fillna(df.age.median())
df.gender = df.gender.fillna('MALE')
df.marital_status = df.marital_status.fillna('Married')
df = df.replace(np.nan, 0)
```

```python
core = pd.get_dummies(data = pd.concat([df[['gender']],df.iloc[:, 4:26], df[['nhomno_xhtdJuin2021','age', 'marital_status_recat']]], axis = 1), columns = ['gender', 'marital_status_recat'], drop_first = True)
core = core.iloc[:, 3:]
core['credible_group'] = np.where(core["nhomno_xhtdJuin2021"].isin([1]), 1, 0)
core = core.drop(['nhomno_xhtdJuin2021'], axis = 1)
core
```

## 2. Model testing

```python
from sklearn.model_selection import train_test_split
X_train_credible, X_test_credible, y_train_credible, y_test_credible = train_test_split(core.drop('credible_group', axis = 1),
                                                                                       core['credible_group'], test_size = 0.2,
                                                                                       random_state = 101)
```

## 2.1. Logistic Regression Model

- GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

logmodel_test = LogisticRegression()
logit_hyper_values = [
    {'C': [0.01, 0.1, 10, 100, 1000, 10000, 100000],
    'solver': ['lbfgs', 'newton-cg', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 1000, 2500, 5000]}
]

find_logit_hyper = GridSearchCV(logmodel_test, param_grid = logit_hyper_values, cv = 4, verbose = True, n_jobs = -1)
find_best_logit_hyper = find_logit_hyper.fit(core.drop('credible_group', axis = 1), core['credible_group'])
best_logit_hyper = find_best_logit_hyper.best_estimator_
```

```python
from sklearn.linear_model import LogisticRegression
logmodel_credible = LogisticRegression(C = 1000, solver = 'newton-cg', class_weight = 'balanced')
logmodel_credible.fit(X_train_credible, y_train_credible)
```

```python
yhat_credible_lr = logmodel_credible.predict(X_test_credible)
```

```python
yhat_credible_lr[:5]
```

```python
logmodel_credible.coef_
```

- Evaluation

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
confusion_matrix(y_test_credible, yhat_credible_lr)
```

```python
print(classification_report(y_test_credible, yhat_credible_lr))
```

```python
logmodel_credible.score(X_test_credible, y_test_credible)
```

## 2.2. Stratified K-Fold Cross-Validation

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
stratified_kf = StratifiedKFold(n_splits = 5)
cvs_score = cross_val_score(logmodel_credible, core.drop('credible_group', axis = 1), core['credible_group'], cv = stratified_kf)
print('Cross Validation Score are {}'.format(cvs_score))
print('Average Cross Validation Score: {}'.format(cvs_score.mean()))
```

## 2.3. K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors = 1, metric = 'minkowski', p = 2, algorithm = 'auto')
knn.fit(X_train_credible, y_train_credible)

yhat_credible_knn = knn.predict(X_test_credible)
```

```python
confusion_matrix(y_test_credible, yhat_credible_knn)
```

```python
print(classification_report(y_test_credible, yhat_credible_knn))
```

## 2.4. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 250, criterion = 'entropy', random_state = 101)
rf.fit(X_train_credible, y_train_credible)
```

```python
yhat_credible_rf = rf.predict(X_test_credible)
```

```python
confusion_matrix(y_test_credible, yhat_credible_rf)
```

```python
print(classification_report(y_test_credible, yhat_credible_rf))
```

## 2.5. Ridge Regression

```python
from sklearn.linear_model import Ridge

r = Ridge()
r.fit(X_train_credible, y_train_credible)

yhat_ridge = r.predict(X_test_credible)

r.score(X_test_credible, y_test_credible)
```

<b> So, K Nearest Neighbors and Random Forest are the chosen methods with highest score.
